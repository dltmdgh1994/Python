```python
import pandas as pd
mydata = pd.read_csv("b.csv",sep=",",encoding = "utf-8")
```

```python
mydata.head()
```

```python
del mydata['영화명']
mydata.head()
```

```python
mydata.dtypes
```

```python
from sklearn.model_selection import train_test_split

y=mydata['1주차누적관객수'].values
x=mydata.drop('1주차누적관객수', axis=1).values

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
```

```python
print(y)
```

```python
#변수 바꾸기 귀찮아서 그냥 이렇게 적어용
train_data_StandardScale = x_train
test_data_StandardScale = x_test
```

```python
#랜덤스테이트 0으로 했는데 왜 계속 바뀌는지 모르겠네요 일단 tree 수 10입니다.
import numpy as np
from sklearn.ensemble import RandomForestRegressor
mlr = RandomForestRegressor(n_estimators = 10, random_state = 0)
mlr.fit(train_data_StandardScale, y_train)
```

```python
print("테스트 세트 정확도 : "+str(mlr.score(test_data_StandardScale, y_test)))
```

```python
print("훈련 세트 정확도 : "+str(mlr.score(train_data_StandardScale, y_train)))
```

```python
#랜덤포레스트 회귀는 특성중요도 해석 가능하답니다!
mlr.feature_importances_
x=['감독파워','배우파워','제작사파워','1주차최대스크린수','개봉전리뷰수','1점비율','2점비율','3점비율','4점비율','5점비율'
            ,'장르_다큐멘터리','장르_드라마',
        '장르_멜로/로맨스','장르_미스터리','장르_범죄','장르_사극','장르_스릴러/공포(호러)',
        '장르_애니메이션','장르_액션','장르_전쟁',
        '장르_코미디','장르_판타지','등급_12세이상관람가','등급_15세이상관람가','등급_전체관람가','등급_청소년관람불가']
j=0
for i in mlr.feature_importances_:
    print(x[j]+" : "+str(round(i,4)))
    j=j+1
```

```python
y_pred = mlr.predict(test_data_StandardScale)
```
