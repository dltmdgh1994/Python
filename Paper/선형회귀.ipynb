```python
from sklearn.model_selection import train_test_split

y=mydata['1주차누적관객수'].values
x=mydata.drop('1주차누적관객수', axis=1).values

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
```

```python
#다중공선성 제거를 위한 정규화
from sklearn.preprocessing import StandardScaler
StandardScaler = StandardScaler()
print(StandardScaler.fit(x_train))
train_data_StandardScaled = StandardScaler.transform(x_train)
test_data_StandardScaled = StandardScaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(train_data_StandardScaled, y_train)
y_pred = mlr.predict(train_data_StandardScaled)
```

```python
print("테스트 세트 정확도 : "+str(mlr.score(test_data_StandardScaled, y_test)))
print("훈련 세트 정확도 : "+str(mlr.score(train_data_StandardScaled, y_train)))
```

```python
import sklearn
from sklearn.metrics import mean_squared_error
```

```python
y_pred = mlr.predict(test_data_StandardScaled)
y_true=y_test
mse=sklearn.metrics.mean_squared_error(y_true, y_pred)
print(int(mse))
print(mse)
```
