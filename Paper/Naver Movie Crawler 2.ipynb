```python
from urllib.request import urlopen
from requests import get
from bs4 import BeautifulSoup
from selenium import webdriver
import re
from pymongo import MongoClient
import time
from operator import itemgetter
import pandas as pd
import numpy as np

hangul = re.compile('[^가-힣|^ |^0-9]+')
driver = webdriver.Chrome('chromedriver')
```

```python
import pandas as pd

def dftoCsv(movie_df, num): 
    try: 
        movie_df.to_csv(('NaverReview'+ str(num) +'.csv'), sep=',', na_rep='NaN', encoding='euc-kr',index = False) 
    except: 
        print("Error")
```

```python
links = []
reviews = []
scores = []
for i in range(1,15):
    page = str(i)
    url = "https://movie.naver.com/movie/sdb/browsing/bmovie.nhn?open=2019&nation=KR&page="+page
    driver.get(url)
    html = driver.page_source
    bs = BeautifulSoup(html,"html.parser")
    for href in bs.find("ul",{"class":"directory_list"}).findAll("li"):
        link = href.find("a")["href"]
        if link[0] == "/":
            links.append(link[28:])
            

for link in links:
    url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code="+link
    driver.get(url)
    html = driver.page_source
    bs = BeautifulSoup(html,"html.parser")
    total = bs.find("strong",{"class","total"}).find("em").text
    total = hangul.sub('',total)
    limit = int(int(total)/10) + 2
    if limit > 2:
        for i in range(1,limit):
            page = str(i)
            url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code="+link+"&page="+page
            driver.get(url)
            html = driver.page_source
            bs = BeautifulSoup(html,"html.parser")   
            try:
                li = bs.find('div',{'class':'score_result'}).findAll('li')
                for l in li:
                    try:
                        review = []              
                        review = l.find('div',{'class':'score_reple'}).findAll("span")
                        review = review[-2].text
                        review = hangul.sub('',review)
                        if not review:
                            continue
                        reviews.append(review)
                        score = l.find('div',{'class':'star_score'}).find("em").text
                        scores.append(score)
                    except:
                        pass
            except:
                pass

movie_dic = {"Review":reviews, 'Score': scores} 
movie_df = pd.DataFrame(movie_dic) 

dftoCsv(movie_df,1)
```
