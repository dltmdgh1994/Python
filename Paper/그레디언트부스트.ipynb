```python
from sklearn.model_selection import train_test_split

y=mydata['1주차누적관객수'].values
x=mydata.drop('1주차누적관객수', axis=1).values

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
```

```python
#여기도 변수 바꾸기 귀찮아서 스케일링 안했지만 걍 써욧!
train_data_StandardScale = x_train
test_data_StandardScale = x_test
```

```python
#그레디언트부스트는 좀 시간오래 걸려요!
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
mlr =  GradientBoostingClassifier(random_state = 0)
mlr.fit(train_data_StandardScale, y_train)
```

```python
print("테스트 세트 정확도 : "+str(mlr.score(test_data_StandardScale, y_test)))
```

```python
print("훈련 세트 정확도 : "+str(mlr.score(train_data_StandardScale, y_train)))
```

```python
#가지치기
mlr =  GradientBoostingClassifier(random_state = 0, max_depth=1)
mlr.fit(train_data_StandardScale, y_train)
print("테스트 세트 정확도 : "+str(mlr.score(test_data_StandardScale, y_test)))
print("훈련 세트 정확도 : "+str(mlr.score(train_data_StandardScale, y_train)))
```

```python
#과적합 방지를 위한 학습률 조절
mlr =  GradientBoostingClassifier(random_state = 0, learning_rate=0.01)
mlr.fit(train_data_StandardScale, y_train)
print("테스트 세트 정확도 : "+str(mlr.score(test_data_StandardScale, y_test)))
print("훈련 세트 정확도 : "+str(mlr.score(train_data_StandardScale, y_train)))
```
